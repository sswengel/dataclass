<p>This is the preliminary data analysis.</p>

<p>First let's load in the tools we will need to analyze this data.</p>

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

<p>Next let's load in the data tables we will be analyzing. We will start with the first data table of earlier years.</p>

```python
df_2002_2011 = pd.read_csv("20022011.csv")
df_2002_2011.head()
```

<p>In this data table, degree awards are sorted by the sector of educational institution, the specific educational institution, the discpline area the degree belongs to, the degree subject, and the level of degree. As for the amount of degrees awarded per year, each year gets its own column.</p>

<p>The sector and discpline area columns are needed for this project, so we can drop them. More importantly, we can define parameters for our data frame that isolate the data to the data we need, that being of bachelor's degrees in history granted by Kent State.</p>

```python
df_2002_2011 = df_2002_2011[df_2002_2011['Institution'] == "Kent State University"][df_2002_2011['Subject Field'] == "History"][df_2002_2011['Level of Degree or Certificate'] == "Bachelor's Degree"]
df_2002_2011 = df_2002_2011.drop(columns=['Sector', 'Discipline Area'])
df_2002_2011
```

<p>The other data frame we will be using has one column for year values as opposed to a column for each year. We can consolidate the year columns in this data frame into the same format as the other data frame using the Melt property.</p>

<p>First, let's rename the year columns to just the year for simplicity's sake.</p>

```python
df_2002_2011 = df_2002_2011.rename(columns={"Awards in FY 2002": "2002",
                                            "Awards in FY 2003": "2003",
                                            "Awards in FY 2004": "2004",
                                            "Awards in FY 2005": "2005",
                                            "Awards in FY 2006": "2006",
                                            "Awards in FY 2007": "2007",
                                            "Awards in FY 2008": "2008",
                                            "Awards in FY 2009": "2009",
                                            "Awards in FY 2010": "2010",
                                            "Awards in FY 2011": "2011"})
df_2002_2011
```

<p>With the Melt feature, we can keep the columns that we want to remain the same as they were (the ID variables) while sublimating the other columns (the value variables) from wide to long format. Then we can rename the default variable and value columns to the actual values they represent</p>

```python
df_2002_2011 = df_2002_2011.melt(id_vars = ['Institution', 'Subject Field', 'Level of Degree or Certificate'], value_vars = ['2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011'])
df_2002_2011
```

```python
df_2002_2011 = df_2002_2011.rename(columns={"variable": "year",
                                            "value": "awards"})
df_2002_2011
```

<hr>

<p>Now let's move on to the other data frame.</p>

```python
df_2012_2023 = pd.read_csv("20122023.csv")
df_2012_2023.head()
```

<p>For this data frame we can do what we did for the other one and isolate out the data that we need for history bachelor's degrees granted by KSU. It does not need Melt at all due to already having a single year column, but it has many more columns that are unnecessary for this analysis and therefore can be cut.</p>

```python
df_2012_2023 = df_2012_2023[df_2012_2023['SUBJECT_TITLE'] == 'History, General'][df_2012_2023['campus'] == 'Kent State University'][df_2012_2023['Degree_Desc'] == 'Bachelor Degree']
df_2012_2023 = df_2012_2023.drop(columns=['sector', 'awlevel', 'UNITID', 'Inst_code', 'subject_field', 'subject_code', 'discipline_area', 'campus_code'])
df_2012_2023
```

<p>While all of this data frame's visible columns have the same values as the first data frame, some of them have different names. Let's rename them so that the two data frames are parallel to each other.</p>

```python
df_2002_2011.columns
```

```python
df_2012_2023 = df_2012_2023.rename(columns={"campus": "Institution",
                                            "SUBJECT_TITLE": "Subject Field",
                                            "Degree_Desc": "Level of Degree or Certificate"})
df_2012_2023
```

<p>Now that the two frames are parallel to each other, we can combine them into one data frame using combine_first.</p>

```python
frame = df_2002_2011.combine_first(df_2012_2023)
frame
```
